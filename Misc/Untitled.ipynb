```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head(5)
```

```python
tips.day.value_counts()
```

```python
# During sunday what is the high amount of tip, if it is given by female or male
highest_tip_sun = tips[tips['day'] == 'Sun'].sort_values(by = ['tip'], ascending = False)
```

```python
highest_tip_sun.head(1)
```

```python
highest_tip_sun.describe() # Highest tip on sunday is 6.5
```

```python
highest_tip_sun['tip'].max() # The tip was given by Male
```

```python
# What is the highest tip given by male & female during sundays
highest_tip_sun_female = highest_tip_sun[highest_tip_sun['sex'] == 'Female']
```

```python
highest_tip_sun_female.sort_values(by = ['tip'], ascending = False) # Highest tip given by female is 5.20
```
